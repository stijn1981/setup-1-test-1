```python
import numpy as np

# Functie om snelheid te corrigeren en maximale waarde te berekenen
def fix_and_get_max_speed(df_speed, naam=""):
    # Stap 1: Zet waarden boven 60 op 0
    df_speed[('measures', 'speed_cm_per_s')] = df_speed[('measures', 'speed_cm_per_s')].mask(
        df_speed[('measures', 'speed_cm_per_s')] > 60, 0)
    
    # Stap 2: Zet 0 naar NaN en interpoleer (overschrijft direct)
    df_speed[('measures', 'speed_cm_per_s')] = df_speed[('measures', 'speed_cm_per_s')].replace(0, np.nan).interpolate()
    
    # Stap 3: Bepaal maximale snelheid
    max_speed = df_speed[('measures', 'speed_cm_per_s')].max()
    
    print(f"Maximale snelheid voor {naam}: {max_speed:.2f} cm/s")
    return max_speed

# Toepassen op de 12 datasets
max_speed1 = fix_and_get_max_speed(df_speed1, "df_speed1")
max_speed2 = fix_and_get_max_speed(df_speed2, "df_speed2")
max_speed3 = fix_and_get_max_speed(df_speed3, "df_speed3")
max_speed4 = fix_and_get_max_speed(df_speed4, "df_speed4")
max_speed5 = fix_and_get_max_speed(df_speed5, "df_speed5")
max_speed6 = fix_and_get_max_speed(df_speed6, "df_speed6")
max_speed7 = fix_and_get_max_speed(df_speed7, "df_speed7")
max_speed8 = fix_and_get_max_speed(df_speed8, "df_speed8")
max_speed9 = fix_and_get_max_speed(df_speed9, "df_speed9")
max_speed10 = fix_and_get_max_speed(df_speed10, "df_speed10")
max_speed11 = fix_and_get_max_speed(df_speed11, "df_speed11")
max_speed12 = fix_and_get_max_speed(df_speed12, "df_speed12")
```

```python
from datetime import timedelta
import numpy as np
import pandas as pd
from scipy.signal import medfilt
def read_timestamps(series):
        return pd.to_datetime(series)

photo_t = read_timestamps(df2["Item1.Timestamp"])
def get_photodiode_triggers(photo, photo_t):
        # First median filter to remove noise
        photo_filt = medfilt(photo, kernel_size=5)
        photo_median = np.median(photo_filt)
 
        below = np.where(photo_filt < 30)[0]
 
        stim_start_idxs = below[np.where(np.diff(below, prepend=0) > 1)[0]]
        stim_end_idxs = below[np.where(np.diff(below, append=99999999) > 1)[0]]
 
        stim_start_times = photo_t.iloc[stim_start_idxs]
        stim_end_times = photo_t.iloc[stim_end_idxs]
 
        # Remove triggers that are shorter than 250ms or longer than 20s apart from each other.
        dt = stim_end_times.reset_index(drop=True) - stim_start_times.reset_index(drop=True)
        to_remove = np.where((dt < timedelta(milliseconds=250)))
        stim_start_idxs = np.delete(stim_start_idxs, to_remove)
        stim_end_idxs = np.delete(stim_end_idxs, to_remove)
        # Remove stimuli within 2 seconods of each other
        stim_start_times = photo_t.iloc[stim_start_idxs]   
        dt = stim_start_times.iloc[1:].reset_index(drop=True) - stim_start_times.iloc[:-1].reset_index(drop=True)
        to_remove = np.where((dt < timedelta(milliseconds=2000)))[0]+1
        stim_start_idxs = np.delete(stim_start_idxs, to_remove)
        stim_end_idxs = np.delete(stim_end_idxs, to_remove)
        print("Number of starts is equal to number of ends:", len(stim_start_idxs)==len(stim_end_idxs))
        triggers = np.zeros((stim_start_idxs.size + stim_end_idxs.size), dtype='uint32')
        triggers[0::2] = stim_start_idxs
        triggers[1::2] = stim_end_idxs
        print(triggers)
        return triggers
```

```python
def get_expa_info(dataframe, triggers, trigger_frames):
    """
    Bepaal de expa_indexes (rij-indexen waar 'EXPA' voorkomt in de kolom 'Value'),
    en haal de bijbehorende **start-triggerwaarden** en **start-framewaarden** uit 'triggers' en 'trigger_frames'.

    Let op: omdat elke stimulus 2 triggers heeft (start & stop), wordt index i → 2*i gebruikt.

    Parameters:
      dataframe      : DataFrame met de stimulusdata (bv. stim1, stim2, ...)
      triggers       : Lijst of array met alle triggerwaarden (start & stop per stimulus)
      trigger_frames : Lijst of array met framewaarden (idem)

    Returns:
      expa_indexes  : Indexen waar 'EXPA' voorkomt in de stimuluslijst
      expa_triggers : Lijst met alleen de start-triggerwaarden (2*i)
      expa_frames   : Lijst met alleen de start-framewaarden (2*i)
    """
    expa_indexes = dataframe.index[dataframe['Value'] == 'EXPA'].tolist()
    expa_triggers = [triggers[2*i] for i in expa_indexes if 2*i < len(triggers)]
    expa_frames   = [trigger_frames[2*i] for i in expa_indexes if 2*i < len(trigger_frames)]
    return expa_indexes, expa_triggers, expa_frames


# Loop over datasets 1 tot en met 12:
for i in range(1, 13):
    # Haal de juiste variabelen op via globals()
    stim = globals()[f"stim{i}"]
    trig = globals()[f"triggers{i}"]
    trig_frames = globals()[f"trigger_frames{i}"]
    
    expa_indexes, expa_triggers, expa_frames = get_expa_info(stim, trig, trig_frames)
    
    # Sla de resultaten op voor later gebruik, indien gewenst
    globals()[f"expa_indexes{i}"] = expa_indexes
    globals()[f"expa_triggers{i}"] = expa_triggers
    globals()[f"expa_frames{i}"] = expa_frames
    
    print(f"Dataset {i}: EXPA indexes: {expa_indexes}")
    print(f"Dataset {i}: EXPA triggers: {expa_triggers}")
    print(f"Dataset {i}: EXPA frames: {expa_frames}")
```

```python
#expa_escape or freeze
import numpy as np

def count_escape_events(speed_data, expa_triggers, threshold=20, frame_count=6):
    """
    Tel escape events in speed_data, gebaseerd op expa_triggers.
    Voor elke trigger (in de lijst expa_triggers) wordt in het 8-seconden venster gecontroleerd of er een blok 
    van 'frame_count' opeenvolgende frames is waarin de snelheid boven het 'threshold'-niveau ligt.
    Als dit zo is, telt die trigger als 1 escape event.
    """
    escape_count = 0
    for start in expa_triggers:
        # Zoek in een venster van 8 seconden na de trigger (60 fps * 8)
        end_index = min(start + 8*60, len(speed_data) - frame_count + 1)
        for i in range(start, end_index):
            if np.all(speed_data[i:i+frame_count] > threshold):
                escape_count += 1
                break  # Voor deze trigger tellen we slechts één escape event
    return escape_count

def count_freeze_events(speed_data, expa_triggers, threshold=2, frame_count=60):
    """
    Tel freeze events in speed_data, gebaseerd op expa_triggers.
    Voor elke trigger wordt in het 8-seconden venster gecontroleerd of er een blok van 'frame_count'
    opeenvolgende frames is waarin de snelheid onder het 'threshold'-niveau ligt.
    Als dit zo is, telt die trigger als 1 freeze event.
    """
    freeze_count = 0
    for start in expa_triggers:
        end_index = min(start + 8*60, len(speed_data) - frame_count + 1)
        for i in range(start, end_index):
            if np.all(speed_data[i:i+frame_count] < threshold):
                freeze_count += 1
                break  # Stoppen als de voorwaarde eenmaal behaald is voor deze trigger
    return freeze_count

# Maak een lijst van (df_speed, expa_triggers) voor datasets 1 t/m 12.
datasets = [(globals()[f"df_speed{i}"], globals()[f"expa_triggers{i}"]) for i in range(1, 13)]

escape_counts = []
freeze_counts = []

for i, (df_speed, expa_triggers) in enumerate(datasets, start=1):
    # Haal de snelheidsextractie (cm/s) als numpy array op
    speed_data = df_speed[('measures', 'speed_cm_per_s')].to_numpy()
    
    escapes = count_escape_events(speed_data, expa_triggers)
    freezes = count_freeze_events(speed_data, expa_triggers)
    
    escape_counts.append(escapes)
    freeze_counts.append(freezes)
    
    print(f"Dataset {i}: Escape events = {escapes}, Freeze events = {freezes}")
```

```python
import numpy as np
import pandas as pd

def detect_freeze_per_trigger(speed_data, expa_triggers, threshold=2, frame_count=60):
    """
    Voor elke trigger controleer of er een freeze-event plaatsvindt in het venster.
    Geeft een lijst terug met 1 (wel freeze) of 0 (geen freeze) per trigger.
    """
    freeze_flags = []
    for start in expa_triggers:
        found_freeze = 0
        end_index = min(start + 8*60, len(speed_data) - frame_count + 1)
        for i in range(start, end_index):
            if np.all(speed_data[i:i+frame_count] < threshold):
                found_freeze = 1
                break
        freeze_flags.append(found_freeze)
    return freeze_flags

# Data verzamelen
all_rows = []

for i, (df_speed, expa_triggers) in enumerate(datasets, start=1):
    speed_data = df_speed[('measures', 'speed_cm_per_s')].to_numpy()
    freeze_flags = detect_freeze_per_trigger(speed_data, expa_triggers)

    mouse_id = (i - 1) // 3 + 1
    session_id = (i - 1) % 3 + 1

    # Shelter-aanwezigheid op basis van jouw regels
    if mouse_id % 2 == 1:
        shelter = 1 if session_id in [1, 3] else 0
    else:
        shelter = 1 if session_id == 2 else 0

    for j, freeze in enumerate(freeze_flags):
        row = {
            'mouse_id': mouse_id,
            'session_id': session_id,
            'trigger_id': j + 1,
            'shelter_presence': shelter,
            'freeze': freeze
        }
        all_rows.append(row)

        # Print elke rij
        print(row)

# Omzetten naar DataFrame
df_long = pd.DataFrame(all_rows)

# Opslaan naar CSV
df_long.to_csv("freeze_trigger_level_expa_test1.csv", index=False)
```

```python
#SWEP20
# Parameters
freeze_threshold = 2
freeze_frames = 60
window_frames = 8 * 60
cm_per_pixel = 0.063
shelter_center_x = 77.0  # Midden van shelter (in cm)

freeze_positions_swep20 = []

for i in range(1, 13):
    df_speed = globals()[f"df_speed{i}"]
    swep20_triggers = globals()[f"swep20_triggers{i}"]
    speed_data = df_speed[('measures', 'speed_cm_per_s')].to_numpy()

    freeze_count = 0

    for start in swep20_triggers:
        end_index = min(start + window_frames, len(speed_data) - freeze_frames + 1)

        has_freeze = False
        for j in range(start, end_index):
            if np.all(speed_data[j:j + freeze_frames] < freeze_threshold):
                has_freeze = True
                break

        if has_freeze and start in df_speed.index:
            x = df_speed['Body']['x'].loc[start] * cm_per_pixel
            y = df_speed['Body']['y'].loc[start] * cm_per_pixel
            distance = abs(x - shelter_center_x)
            freeze_count += 1
            freeze_positions_swep20.append({'dataset': i, 'x': x, 'y': y, 'distance': distance})
            print(f"Dataset {i} — Freeze {freeze_count}: x = {x:.2f} cm, y = {y:.2f} cm → afstand = {distance:.2f} cm")
        elif has_freeze:
            print(f"Dataset {i} — Freeze skipped (trigger frame {start} not in index)")

# Optioneel: gemiddelde afstand tonen
if freeze_positions_swep20:
    avg_distance = sum(p['distance'] for p in freeze_positions_swep20) / len(freeze_positions_swep20)
    print(f"\nGemiddelde afstand tot shelter (SWEP20 FREEZE): {avg_distance:.2f} cm")
else:
    print("\nGeen SWEP20 freezes gevonden.")
```
